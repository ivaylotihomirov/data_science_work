```python
import pandas as pd

data = pd.read_csv('/Users/ivaylo.ivanov/Downloads/Guided Project_ Predicting the stock market (1)/sphist.csv')

print(data.head())
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date']) 
data['Day of week'] = data['Date'].dt.dayofweek

data.info()
```

```python
from datetime import datetime

data['before 2013'] = data['Date'] < datetime(year=2013,month=1,day=1)

data.sort_values(['Date'], ascending=False)
```

```python
data['rolling avg 5'] = data['Close'].rolling(window=5, center=False, min_periods=1).mean()
data['rolling avg 30'] = data['Close'].rolling(window=30, center=False, min_periods=1).mean()
data['rolling avg 365'] = data['Close'].rolling(window=365, center=False, min_periods=1).mean()
data['volume avg 5'] = data['Volume'].rolling(window=5, center=False, min_periods=1).mean()
```

```python
data.sort_values(['Date'], ascending=False)
```

```python
train = data[data['before 2013']==True].copy()
test = data[data['before 2013']==False].copy()
```

```python
train.head()
```

```python
test.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

train_features = train[['rolling avg 5','Day of week','volume avg 5']]
train_target = train['Close']

test_features = test[['rolling avg 5','Day of week','volume avg 5']]
test_target = test['Close']

lr = LinearRegression()

lr.fit(train_features, train_target)

predictions = lr.predict(test_features)

mae = mean_absolute_error(predictions, test_target)

print('The MAE is', mae)
```

There is not much of a reduction in the error, when additional variables apart from rolling 5 day average price have been included.
