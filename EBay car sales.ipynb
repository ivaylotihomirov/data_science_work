Converting both columns to integers

```python
autos["price"] = autos["price"].str.replace("$","").str.replace(",","").astype(int)
print(autos["price"].head())
autos["odometer"] = autos["odometer"].str.replace(",","").str.replace("km","").astype(int)
print(autos["odometer"].head())
```

# Examine outliers in the price column

```python
print(autos["price"].value_counts().head())
print("\n")
print(autos["price"].value_counts().sort_index(ascending=True))
```

Some of the prices do not seem to be realistic. There are 1,421 cars with 0 price, and some cars have prices in the millions.

I will remove all cars with values with values higher than 10000000 USD as there are very few such vehicles and also vehicles with price of $0 (something sold/bought should always have a value). The issue with removing $0 vehicles is that it is the most frequent category.

```python
autos = autos[autos["price"].between(1, 9999999)]
print(autos["price"].value_counts().sort_index(ascending=True))
```

# Exploring unique brands

```python
unique = autos["brand"].unique()
print(unique)
```

```python
top_6 = autos['brand'].value_counts()[0:6]
```

```python
print(top_6.index)
```

```python
price_dict = {}

for c in top_6.index:
    selection = autos.loc[autos["brand"]== c,"price"].mean()
    price_dict[c] = selection

print(price_dict)
```

BMW, Audi and Mercedes have higher mean prices than Opel, Volkswagen and Ford. This does not definitively imply that these brands have higher value vehicles (although this would make sense), because we are not controlling for registration year and type of vehicle.

```python
# creating a new dictionary for mileage #

mileage_dict = {}

for c in top_6.index:
    selection = autos.loc[autos["brand"]== c,"odometer_km"].mean()
    mileage_dict[c] = selection

print(mileage_dict)
```

Creating series

```python
price = pd.Series(price_dict)
mileage = pd.Series(mileage_dict)
print(price)
print('\n')
print(mileage)     
```

```python
df = pd.DataFrame(price,columns = ["price"])
print(df)
```

```python
df["mileage"] = mileage
```

```python
print(df)
```

It looks like mileage on the brands is comparable - this means that we have more ground for stating that BMW, Audi and Mercedes' vehicles truly have higher values than vehicles by VW, Opel and Ford.
